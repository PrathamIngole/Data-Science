            - name attribute

```python
import pandas as pd
import numpy as np 

s = pd.Series([100, 'rossum', 34.54, 'Distinction'], index = ['ID', 'Name', "percentile", "status"], name = "Student")
print(s, type(s))
# name will give the name of the series.
```

```python
#we can get name of series object by using seriesobj.name
print("series name : ", s.name)
```

```python
# we can also change the name of the series object  :
s.name = "stdnt"
print(s)
```

        - dimension

```python
# find the dimension, shape by using ndim and shape attribute
print("dimension of series object : ", s.ndim)
print("Shape of series object : ", s.shape)
```

        - size

```python
# find the number of values present in Series object (includes None, NaN, NA) :

s = pd.Series([1, 45, None, "rustum", pd.NA, 4.6, np.nan], name="altu faltu")
print("size of thee series : ", s.size)
print(s)
print("number of non None values : ", s.count())
print("number of None values : ", s.size - s.count())
```

        - indices of the series

```python
# when we don't give the index of series then it will give us the range
s = pd.Series([1, 45, None, "rustum", pd.NA, 4.6, np.nan], name="altu faltu")
print("size of thee series : ", s.size)
print(s)
print("index of the series ")
print(s.index)
```

```python
# when we give the index of the series when initialization 
s = pd.Series([100, 'rossum', 34.54, 'Distinction'], index = ['ID', 'Name', "percentile", "status"], name = "Student")
print("size of thee series : ", s.size)
print(s)
print('series indices is : ')
print(s.index)
```
