## Number of ways to create of object of series 

+ fundamental data type of python.

```python
import pandas as pd
a = 199
print("type of a : ", type(a))

s = pd.Series(a)
print("content of series : ")
print(s, type(s))
```

```python
import pandas as pd
a = 199
print("type of a : ", type(a))

s = pd.Series(a, dtype=float)
print("content of series : ")
print(s, type(s))
```

```python
import pandas as pd
a = 234.6
print("type of a : ", type(a))

s = pd.Series(a)
print("content of series : ")
print(s, type(s))
```

```python
import pandas as pd
a = True
print("type of a : ", type(a))

s = pd.Series(a)
print("content of series : ")
print(s, type(s))
```

```python
import pandas as pd
a = 2 + 3.5j
print("type of a : ", type(a))

s = pd.Series(a)
print("content of series : ")
print(s, type(s))
```

+ Sequential data type of python.

```python
import pandas as pd
a = 'Pratham'
print("type of a : ", type(a))

s = pd.Series(a)
print("content of series : ")
print(s,"\n", type(s))
```

```python
import pandas as pd
r = range(10, 21, 2)
print("type of r : ", type(r))

s = pd.Series(r)
print("content of series : ")
# series will give default index of the elements (0 based indexing)
print(s,"\n", type(s))
```

+ List Data Types

```python
import pandas as pd
lst = [10, 21, 2, 344, 532, 20]
print("type of lst : ", type(lst))

s = pd.Series(lst)
print("content of series : ")
print(s,"\n", type(s))
```

```python
import pandas as pd
lst = [10, 21, 2, 344, 532, 20]
print("type of lst : ", type(lst))

# user difined indices
s = pd.Series(lst, dtype=float, index=[1,2,3,4,5,6])
print("content of series : ")
print(s,"\n", type(s))
```

```python
import pandas as pd
lst = [10, 21, 2, 344, 532, 20]
print("type of lst : ", type(lst))

# user defined indices : 
s = pd.Series(lst, index=("a", "b", "c", "d", "e", "f"))
print("content of series : ")
print(s,"\n", type(s))
```

```python

import pandas as pd
lst = [10, 21, 2, 344, 532, 20]
print("type of lst : ", type(lst))

# for large data sets 
s = pd.Series(lst, index= [f"val {i+1} is --> "for i in range(len(lst))])
print("content of series : ")
print(s,"\n", type(s))
```

+ Hetrogenous type of sequence in series : 

```python

import pandas as pd
lst = [10, "hey there", 2.50, 34 + 4j, True]
print("type of lst : ", type(lst))

# for large data sets 
s = pd.Series(lst, index= [f"{i+1}"for i in range(len(lst))])
print("content of series : ")
print(s,"\n", type(s))
```

#### Creating the series object using the dict object.

```python

import pandas as pd
d = {
    "python" : 1, 
    "C++" : 2, 
    "Java" : 3, 
    "DSA" : 4
}
print("type of d : ", type(d))

# for large data sets 
# when dictionary is passed as a data set then the keys become the indices and the value became the series data.
s = pd.Series(d)
print("content of series : ")
print(s,"\n", type(s))
```

+ Creating Series uning ndarray type

```python
import pandas as pd
import numpy as np

arr = np.eye(3, 3, dtype=int).flatten()
arr2 = np.array([i for i in range(5, 11)])
print(arr, type(arr))

s = pd.Series(arr) 
s2 = pd.Series(arr2, index = [i+1 for i in range(len(arr2))])
print("content of the series s : ")
print(s)
print(type(s))
print("content of the series s2 : ")
print(s2)
print(type(s2))
```
