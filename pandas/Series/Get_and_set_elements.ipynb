### Get and Set series elements :

```python
import numpy as np
import pandas as pd
```

```python
series = pd.Series(np.arange(45, 56), index=[f"ID{i}" for i in range(45, 56)], name = "faltu Series")
```

```python
series.size
```

### Getting values

```python
print(series.size)
print(series)
series["ID56"] = 54554
print(series.size)
print(series)
```

### Setting / Changing / Updating values 


```python
import numpy as np
import pandas as pd

s = pd.Series(np.array(["Srinivas", "Pratik", "Raj", "Rossum", "Kinny", "Rajesh"]))

```

```python
# updating the value of the index 2 :

s[2] = "Rajkumar"
s
```

```python
# setting the values of the series of the using the silicing : 

s[3:] = ["riddhi", "piddu", "sam"]
s
```

```python
# updating random multiple values by using random indices   of series Object

s[[1, 2, 4]] = ["pratham", "vashu", "namam"]
s.size
```

```python
s.index = [f"Name{i+1}" for i in range(s.size)]
s
```

```python
s["Name3"] = "Ingole"
```
