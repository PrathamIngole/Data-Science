#### ***Task = 1 :  frequency table development without using value_count()***

```python
data = {
    
}

for val in visa["continent"] :
    if val in data :
        data[val] += 1 
    else :
        data[val] = 0 

pd.DataFrame([data.keys(), data.values()], index = ["continent", "frequencies"]).T.sort_values(by = "continent")
```

#### ***Task-2: Create a Data frame from Value counts***

```python
pd.DataFrame([visa.continent.value_counts().index, visa.continent.value_counts().values], index = ["continent", "frequencies"]).T
```

#### ***Task-3:  Create a Frequency table using Group By***


```python
pd.DataFrame([visa.groupby(by= "continent").count()["case_id"].index, visa.groupby(by= "continent").count()["case_id"].values], index = ["continent", "frequency"]).T
```

#### ***Task-4:  Write all categorical frequency tables in a new folder***

```python
folder = "Task-4"
os.makedirs(folder, exist_ok=True)
path = os.path.join(os.getcwd(), folder)
```

```python
for i in cat[1:] : 
    pd.DataFrame([visa[i].value_counts().index, visa[i].value_counts().values], index = [i, "frequency"]).T.to_csv(f'{path}\\{i}.csv')
```

#### ***Task-5: Implement a Describe function with using method for Numerical columns***
