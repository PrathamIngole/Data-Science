## ***Feature Scaling***


- Feature Scaling is a method, to modify the value either increase  or decrease.

- In ML we have different columns

        + age range between 0 to 100

        + height range between 0 to 200

        + income range between 10k to 1Crore

- differrent columns having different range of values also different units

- ML model might be think more values leads to more impact 

- scale all the columns into a one scale is a better approach when model are works more on numbers.

- Distance based algorithm are need to be scale

- Deep learning main aim is lower the error : ***Convergence***


--> Two Types of scaling :

        - 1. Standard Scalar

## ***Steps***

- Step - 1 : read the previlaing wage data : wage_data(x)

- Step - 2 : calculate mean

- Step - 3 : calculated std 

- Step - 4 : Nr = s1 - s2

- Step - 5 : Nr / s3

```python
visa = pd.read_csv(r"C:\Users\prath\FullStackDataScience\my own folder\database\data_files\Visadataset.csv")
visa.head(3)
```

```python
visa['prevailing_wage']
```

```python
mean = visa['prevailing_wage'].mean()
```

```python
std = visa['prevailing_wage'].std()
```

```python
Nr = visa['prevailing_wage'] - mean
```

```python
Nr
```

```python
Z_score = Nr / std
```

```python
Z_score
```

```python
plt.figure(figsize = (14, 10))
plt.title("comparision between scaled and original data")
plt.subplot(2, 2, 1).hist(visa['prevailing_wage'])
plt.subplot(2, 2, 2).hist(Z_score)
plt.subplot(2, 2, 3).boxplot(visa['prevailing_wage'])
plt.subplot(2, 2, 4).boxplot(Z_score)

plt.show()
```
