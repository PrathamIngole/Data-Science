### ***Encoding***

            + Encoding means convert categorical data Numerical data 
            + ML/DL or any model works based on maths 
            + Models can not understand english caharaters 
            + so it is very important to convert categorical label to numerical values
            + For examples Gender columns has two tables

                        - Male 
                        - Female

            + Then Female represent as 0 and male represent as 1 based on alphabetical order.

```python
## visa[case_status]

visa.case_status.unique()
```

```python
d = {
    "Certified" : 0,
    "Denied" : 1

}

visa["case_status"] = visa["case_status"].map(d)
```

```python
visa
```

```python
## for continent column :

con = visa.continent.unique().tolist()

continent = { con[i] : i  for i in range(len(con))}
```

```python
visa['continent'] = visa['continent'].map(continent)
```

```python
visa
```

```python
l = visa.continent.unique().tolist()
visa["continent"] = visa["continent"].map({l[i] : i for i in range(len(visa.continent.unique().tolist()))})
```
