## ***Numerical v/s Numericacl***

            - plt.scatter

```python
x = np.arange(11)
y = np.arange(11)

plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

```python
x2 = np.arange(-4, 5)

plt.scatter(x2, x2**2)
plt.plot(x2, x2**2)
plt.show()
```

```python
num = visa.select_dtypes(exclude= "str").columns.tolist()
num
```

```python
col = visa[num[0]]
col2 = visa[num[1]]
col3 = visa[num[2]]

plt.figure(figsize = (14, 5))

plt.subplot(1, 3, 1).scatter(col, col2)
plt.title("no of employees vs yr of estab")
plt.subplot(1, 3, 2).scatter(col2, col3)
plt.subplot(1, 3, 3).scatter(col3, col)
plt.suptitle("Scatter plot")
plt.show()
```
