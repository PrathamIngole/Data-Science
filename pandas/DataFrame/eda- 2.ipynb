```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
visa = pd.read_csv(r"C:\Users\prath\FullStackDataScience\my own folder\database\data_files\Visadataset.csv")
visa.dtypes
categorical = visa.select_dtypes(include = "str").columns
numerical = visa.select_dtypes(exclude = "str").columns
```

```python
folder = "Categorical_Columns_Analysis"
os.makedirs(folder, exist_ok=True)
path = os.path.join(os.getcwd(), folder)
path
```

```python
col = ["red", "blue", "green", 'Yellow', "Orange", "Pink"]

for i in range(1, len(categorical)) :
    cate = visa[categorical[i]].value_counts().index
    frequency = visa[categorical[i]].value_counts().values
    df = pd.DataFrame({f"{categorical[i]}" : cate, f"frequency_of_{categorical[i]}" : frequency})
    newPath = os.path.join(path, f"{categorical[i]}")
    # df.to_csv(f"{newPath}.csv")
    plt.Figure(figsize=(9, 6))
    plt.bar(
        cate,
        frequency,
        color = [col[np.random.choice(6)] for i in cate]
    )
    plt.xlabel(f"class : {i}")
    plt.ylabel(f"frequency of {i}")
    plt.title(f"{i} class / frequency chart")
    # plt.savefig(f"{newPath}.jpg")
    plt.show()
```

```python
keys = visa["continent"].value_counts().keys()
columns = visa["continent"].value_counts().values

col = ["red", "blue", "green", 'Yellow', "Orange", "Pink"]

plt.Figure(figsize=(9, 6))
plt.bar(
    keys, 
    columns,
    color = col
)
plt.xlabel("Class")
plt.ylabel("NO_of_Applicants")
plt.title("continent_bar_chart")
plt.show()
```

```python
folder = "PieCharts"
# os.makedirs(folder)
path = os.path.join(os.getcwd(), folder)


for i in categorical[1:] :
    lab = visa[i].value_counts().index
    values = visa[i].value_counts(normalize = True).values
    newPath = os.path.join(path, f"{i}.png")
    plt.pie(
        values,
        explode=None,
        labels=lab,
        autopct= "%0.2f%%",
        radius = .7,
    )
    plt.title(f"{i}_chart")
    plt.show()
    # plt.savefig(newPath)
```
