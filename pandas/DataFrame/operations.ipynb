# *Step - 12* : Take two dataframes df1 and df2 : 

            - same columns
            - df1 is having two rows 
            - df2 is having two rows 

#### pd.concat([df1, df2])

```python
data1 = {
    "name" : ["pratham"],
    "qualification" : ["B.Tech"],
}
data2 = {
    "name" : ["rohit"],
    "qualification" : ["B.Com"],
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
pd.concat([df1, df2], ignore_index= True).reset_index()
```
