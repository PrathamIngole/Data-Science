# How to add new columns in data frame :

            -- df["New COlumn"] = ["elents]

```python
student_data = {
    "htno": [100, 101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111],
    "name": ["Ramesh", "Rajesh", "Rossum", "Raji", "Kalyan", "Karthik",
             "Kambli", "Praveen", "Ganesh", "Nags", "Biswa", "Ritchi"],
    "telugu": [50, 45, 56, 56, 51, 48, 53, 46, 53, 55, 66, 66],
    "english": [60, 67, 88, 78, 63, 62, 81, 88, 62, 77, 48, 68],
    "hindi": [66, 34, 56, 34, 62, 39, 59, 74, 76, 44, 86, 64],
    "maths": [98, 67, 99, 56, 93, 68, 92, 86, 88, 77, 95, 76],
    "science": [66, 66, 44, 88, 67, 65, 48, 78, 76, 86, 48, 98],
    "social": [55, 78, 77, 55, 51, 88, 73, 45, 35, 58, 47, 75]
}

students = pd.DataFrame(student_data)

print(students)
```

```python
students["total_marks"] = students["telugu"] + students["english"] + students["hindi"] + students["maths"] + students["science"] + students['social']
```

```python
students
```

```python
students["percentage"] = (students["total_marks"] / 6).round(2)
```

```python
students
```

```python
students["percentage"].round(1)
```

```python
students
```

```python
# setting the grades of the students :
# one method -- there are others as well.

students["grades"] = ["A" if i > 70 else "B" if i > 60 else "C" if i > 50 else "F" for i in students["percentage"]]
```

```python
students
```

# data filtering with conditions and updations :

```python
students
```

```python
del students["grades"]
```

```python
# again adding grades 
students["grades"] = None
students
```

```python
# Precentage ---> grade :

# 100 - 75 --> Distinctions 
# 75 - 65 -->   First
# 65 - 55 --> Second
# 55 - 33 --> third
# <33 ---> Fail

students.loc[(students["percentage"] <= 100) & (students["percentage"] >= 75), ["grades"]] = "DISTINCTION"
students.loc[(students["percentage"] < 75) & (students["percentage"] >= 65), ["grades"]] = "First"
students.loc[(students["percentage"] < 65) & (students["percentage"] >= 55), ["grades"]] = "Second"
students.loc[(students["percentage"] < 55) & (students["percentage"] >= 33), ["grades"]] = "Third"
students.loc[students["percentage"] < 33, ["grades"]] = "Fail"
students
```
