```python
# Problem-1:
# From the raw data below create a Pandas Series 'Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat'
# a) Print all elements in lower case
# b) Print all the elements in upper case
# c) Print the length of all the elements

import pandas as pd
import numpy as np 

lst = ['Amit', 'Bob', 'Kate', 'A', 'b', np.nan, 'Car', 'dog', 'cat']
s = pd.Series(lst)
print("series is : \n" ,s)
print("---------------------------------------------------")
print("lower case : ")
lower = pd.Series([v.lower() if isinstance(v, str) else v for v in s ])
for i in lower :
    print("\t", i)
print("---------------------------------------------------")
print("lower case : ")
upper = pd.Series([v.upper() if isinstance(v, str) else v for v in s ])
for i in upper :
    print("\t", i)
print("---------------------------------------------------")
print("length of the each element of the series : ")
for val in s : 
    if isinstance(val, str):
        print(len(val))
    else :
        print(val)
    
```

```python
# 6.2: From the raw data below create a Pandas Series 
# ' Atul', 'John ', ' jack', 'Sam'
# a) Print all elements after stripping spaces from the left and right
# b) Print all the elements after removing spaces from the left only
# c) Print all the elements after removing spaces from the right only
```

```python
s2 = pd.Series([" Atul", "John ", " jack", "Sam"])
for v in s2 :
    if type(v) == str : 
        print(f"\t{v.strip()}")
    else :
        print(v)
```

```python
s2 = pd.Series([" Atul", "John ", " jack", "Sam"])
for v in s2 :
    if type(v) == str : 
        print(f"\t{v.lstrip()}")
    else :
        print(v)
```

```python
s2 = pd.Series([" Atul", "John ", " jack", "Sam"])
for v in s2 :
    if type(v) == str : 
        print(f"\t{v.rstrip()}")
    else :
        print(v)
```
