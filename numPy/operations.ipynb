## Broadcasting (Why NumPy is Powerful)
            - Broadcasting allows NumPy to perform operations on arrays of different shapes.

```python
arr2 = np.arange(1, 5)
print(f"array 2 is : \n\n{arr2}\n\n array 2 if multiply with 2 then (arr2 * 2) : \n\n\n{arr2 * [3, 5, 1, 2]}")

```

```python
arr2 = arr2.reshape((2, 2))
print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand its multiplication or any other operation does the same effect : \n\n{arr2*10}")
```

#### if you are giving single elemnt into broadcasting then it will do operation into each elements into the array else if youre giving some list structure to do operations that should be the same shape as numpy array else it will throw error

```python
print(f"array 2 is : \n\n{arr2}\n\n array 2 if multiply with 2 then (arr2 * 2) : \n\n\n{arr2 * [3, 5, 2]}")
```

## Aggregation Functions : 

            --> like 

                    1. np.sum(arr)	    -->     Total sum
                    2. np.mean(arr)	    -->     Average
                    3. np.max(arr)	    -->     Max
                    4. np.min(arr)	    -->     Min
                    5. np.std(arr)	    -->     Standard deviation

```python
print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}")
```

```python
sum_of_arr2 = np.sum(arr2)

print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand the sum of all elements is : {sum_of_arr2}")
```

```python
sum_col_wise = np.sum(arr2, axis=0)
print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand the sum of all elements Column wise is : {sum_col_wise}")
```

```python
sum_row_wise = np.sum(arr2, axis=1)
print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand the sum of all elements Row wise is : {sum_row_wise}")
```

+ mean of array 


```python
mean = np.mean(arr2)
print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand the mean of all elements is : {mean}")
```

```python
mean_Col_wise = np.mean(arr2, axis=0, dtype=int)
print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand the mean of elements Column wise is : {mean_Col_wise}")
```

```python
mean_row_wise = np.mean(arr2, axis=1, dtype=float)
print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand the mean of elements row wise is : {mean_row_wise}")
```

+ max and min 

```python
maxi = np.max(arr2)
mini = np.min(arr2)

print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand minimum elements is : {mini}\n\nand maximum element is : {maxi}")
```

```python
maxi_row_wise = np.max(arr2, axis=1)
mini_row_wise = np.min(arr2, axis=1)

print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand minimum elements in each row is : {mini_row_wise}\n\nand maximum element in each row is : {maxi_row_wise}")
```

```python
maxi_col_wise = np.max(arr2, axis=0)
mini_col_wise = np.min(arr2, axis=0)

print(f"new reshaped(2, 2) arr2 is : \n\n{arr2}\n\nand minimum elements in each col is : {mini_col_wise}\n\nand maximum element in each col is : {maxi_col_wise}")
```
