                + Lists

                + tensors


        -  List and Tuples are basic level representatitons with no shape and dimensions.
        -  Math operations are slow with List and Tuples.
        -  Numpy arrays are having shapes and dimensions.
        -  Numpy arrays and Tensors are both same in nature.
        -  When CPU is there Numpy arrays are enough to perform the operations.
        -  When GPU is there tensors are required to perform the operations.

+ **Why Numpy**

```python
l1 = [10, 20, 30, 40]
l2 = [100, 200, 300, 400]

ans = []

i = 0 
j = 0 

while i < len(l1) and j < len(l2) :
    ans.append(l1[i] + l2[j])
    l1[i] += l2[j]
    i += 1
    j += 1

l1, ans
```

```python
array = np.array(l1)
array2 = np.array(l2)
```
