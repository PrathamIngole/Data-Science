-- statictical method by numpy

```python
import numpy as np 
```

```python
def row_cumsum(lst) :
    res =  [[0 for i in range(len(lst[0]))] for i in range(len(lst))]
    for i in range(len(lst)) :
        sum = 0 
        for j in range(len(lst[0])) : 
            sum = sum + lst[i][j]
            res[i][j] = sum
    return res
```

```python
lst = [
    [11, 2, 33],
    [4, 15, 6],
    [7, 8, 9]
]

row_cumsum(lst)
```

```python
np_arr = np.array(lst) 

np_arr.cumsum(axis = 1)
```

```python
np.argmax(lst, axis= 0)
```
