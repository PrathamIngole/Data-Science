## Variance : 
                           / sumation (x - u)Â²\
                 VAR  =   | -----------------  |
                           \     N            /

     - step 1. calculate the mean.
     - step 2. substract every element from mean.
     - step 3. square of all the step2 elements
     - step 4. calculate numerator sum of all the element of step 3.
     - step 5. calculate denominator i.e. size of array.
     - step 6. calculate variance i.e. numerator / denominator.

```python
print("original array : ", arr)
mean  = np.mean(arr)
print("mean : ", mean)
arr2 = arr - mean
print("step 2 x - u : ", arr2)
arr3 = arr2 ** 2
print("square all : ", arr3)
Nm = np.sum(arr3)
print("summation of all (Numerator) : ", arr4)
Dm = np.size(arr)
print("N : ", Dm)
var = Nm / Dm
print("variance : ", var)
```

### Co-Variance

                        cov (x, y) =  /   summation (x - mean(x))(y - mean(y)) \
                                     |   ------------------------------------   |
                                      \                  N                     /

    

```python
arr1 = np.arange(10, 21)
arr2 = np.arange(20, 31)

mean_x = np.mean(arr1)
mean_y = np.mean(arr2)

A = arr1 - mean_x
B = arr2 - mean_y

arr3 = np.multiply(A , B)

Nu = np.sum(arr3)

cov = Nu / np.size(arr3)
cov
```

```python
arr11 = np.arange(10, 21)
arr12 = np.arange(20, 31)

np.sum(np.multiply((arr11 - np.mean(arr11)),(arr12 - np.mean(arr12)))) / np.size(arr11)
```

```python
A = np.arange(10, 21)
B = np.arange(20, 31)
cov_x_y = np.cov(A, B)
cov_x_y
```
