so we already have the columns for step 12 - 15, lets first delete it and the we can again do that from the scratch.

```python
# lets check the columns first 
students.columns
```

```python
# lets delete multiple rows in one go. 

new_students = students.drop(["total_marks", "percentage", "grades"], axis = 1)
```

```python
# checking the dataframe 
print(students)
print(new_students)
```

```python
# lets move towards the step 12 as now we want total marks :

new_students["total_marks"] = students[["telugu","english","hindi", "maths", "science","social"]].sum(axis= 1)
new_students
```

```python
# lets define new percentage column

new_students["percentage"] = round(new_students["total_marks"]/6, 2)
new_students
```

```python
# lets define the grade column 

# we have two ways 1st :

new_students["grade"] = [ "DISTINCTION" if i >= 75 else "FIRST" if i >= 65 else "SECOND" if i >= 55 else "THIRD" for i in new_students["percentage"]]
new_students
```

```python
# for second way first we need to drop the grade column
new_students.drop(["grade"], axis=1, inplace = True)
new_students
```

```python
# second way to do it 

# we will define a user defined function for this task :
def set_grade(percent) :
    if percent >= 75 :
        return "DISTINCTION"
    elif percent >= 65 :
        return "FIRST"
    elif percent >= 55 :
        return "SECOND"
    else :
        return "THIRD"
    

new_students["grades"] = new_students["percentage"].apply(set_grade)
new_students
```

```python
# now we see the aggregation of all the subjects 

average_marks = new_students[["telugu", "english", "hindi", "maths", "science", "social"]].mean(axis=0)
average_marks
```
