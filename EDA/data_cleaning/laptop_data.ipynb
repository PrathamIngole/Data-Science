#### *****Let's create a table Cpu category which Specify The issue is intel, AMD Or Samsung.*****

            - For that we need to write a function.

            - Which takes each value of the Cpu Column And classify it whether it is a intel AMD Or Samsung.

```python
def cpu_cat(x) :
    if "Intel" in x :
        return "Intel"
    elif 'AMD' in x : 
        return "AMD"
    elif "Samsung" in x :
        return "Samsung"
```

```python
df["cpu_cat"] = df["Cpu"].apply(cpu_cat)
```

```python
df.info()
```

```python
df.head()
```

```python
df.cpu_cat.value_counts()
```

```python
# now lets label as 0 , 1, 2 for model as machine only takes numerical columns : 

def label_cpu(x) :
    if x == "Intel" :
        return 0 
    elif x == "AMD" :
        return 1 
    else :
        return 2 
```

```python
df['cpu_cat'] = df['cpu_cat'].apply(label_cpu)
```
