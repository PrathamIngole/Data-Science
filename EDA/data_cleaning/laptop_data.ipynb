```python
# Now let's change Weight DType

df["Weight"] = df["Weight"].str.replace("kg", "").astype(float)

## here "Weight" is our columns contains floating values + kg in its positon. 
            # so (.str) funciton changes its values into string.
            # then (.replace(this, to-this)) function will change the "kg" to "" --> empty string. 
            # then at last (.astype ) will change and redefine its values type to "float".
```

#### ***Here we can observe***

        - there There is so many information in one column here 
        - So we need to split that information into separate parts.
        - which is IPS_panel 
        - Second one TouchSscreen 
        - Third one is a resolution Which is :
                    
                    1. x_res and
                    2. y_res.

        - So let's make Three columns.

```python
## first column is 

df["IPS_panel"] = df["ScreenResolution"].apply(lambda x : 1 if "IPS" in x else 0)
```

```python
df.head()
```

```python
## lets make second columns :

df["TouchScreen"] = df["ScreenResolution"].apply(lambda x : 1 if "Touchscreen" in x else 0)
```

```python
df.head()
```

```python
## lets work on the resolutions 

string = 'IPS Panel Retina 43 x 45 Display 2560x1600'
pattern = "\d+x\d+"

re.findall(pattern, string) # Now we use this to select the new columns 
```

```python
df["x_y_res"] = df["ScreenResolution"].apply(lambda x : re.findall(pattern, x)[0])
```

```python
df.head()
```

```python
# now seperate the x_res and the y_res

df["x_res"] = df["x_y_res"].apply(lambda x : x.split('x')[0])
df["y_res"] = df["x_y_res"].apply(lambda x : x.split('x')[1])

# here we seperate thenm both
```

```python
df.head()
```

```python
df.info()
```

```python
## noticed something we have placed all three columns in the DATAFRAME but can you observe that the Dtype of all the three colunms is str but we need x_res and y_res into numerical values (int).

# so ....

df["x_res"] = df["x_res"].astype(int)
df["y_res"] = df["y_res"].astype(int)
```

```python
df.info()
```

#### *****Let's create a table Cpu category which Specify The issue is intel, AMD Or Samsung.*****

            - For that we need to write a function.

            - Which takes each value of the Cpu Column And classify it whether it is a intel AMD Or Samsung.

```python
def cpu_cat(x) :
    if "Intel" in x :
        return "Intel"
    elif 'AMD' in x : 
        return "AMD"
    elif "Samsung" in x :
        return "Samsung"
```

```python
df["cpu_cat"] = df["Cpu"].apply(cpu_cat)
```
