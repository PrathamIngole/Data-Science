```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\prath\FullStackDataScience\Data-Science\EDA\data_sets\stroke-data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns = ["id"], inplace = True)
```

```python
df["gender"].value_counts()
```

```python
df['gender'] = df['gender'].apply(lambda x : 0 if x == "Female" else 1 )
```

```python
df.info()
```

```python
df["ever_married"].value_counts()
```

```python
df["ever_married"] = df["ever_married"].apply(lambda x : 1 if x == "Yes" else 0)
```

```python
df['work_type'].value_counts()
```

```python
df["work_type"]  = df["work_type"].map({"Private" : 0, "Self-employed" : 1, "children" : 2
, "Govt_job" : 3, "Never_worked" : 4
})
```

```python
df["work_type"].value_counts()
```

```python
df["Residence_type"].value_counts()
```

```python
df["Residence_type"] = df["Residence_type"].apply(lambda x : 0 if x == "Urban" else 0 )
```

```python
df["bmi"] = df["bmi"].fillna(np.mean(df["bmi"]))
```

```python
df["smoking_status"].value_counts()
```

```python
df["smoking_status"] = df["smoking_status"].map({"never smoked" : 0, "Unknown" : 1, "formerly smoked" : 2, "smokes" : 3})
```
