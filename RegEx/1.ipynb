### **re.search**
    - searches the pattern for anywhere in a given string 

```python
pattern = "hi"
str1 = "how hi are hi you"
result = re.search(pattern, str1)
print("result : ", result)
result.group()
```

### **re.findall**
        - using a \d+ pattern 
        - \d indicates digit match between 0 - 9 
        - plus indicates one or more occurences of digit

```python
## using normal method : 

str1 = "the price is 100 and the tax is 20"

lst = str1.split()

for i in lst :
    if i == '100' :
        print(i)
        break
```

```python
## using the re.findall
pattern = r"\d+"

str1 = "the price is 100 and the tax is 20"

result = re.findall(pattern, str1)
print("result : ", result)
```

```python
## using the re.findall
pattern = r"\d+"

str1 = "order number 1234 shipped on 28th December 2024"

result = re.findall(pattern, str1)
print("result : ", result)
```

```python
## using normal method : 

str1 = "order number 1234 shipped on 28th December 2024"

lst = str1.split()

for i in lst :
    for j in i :
        if j.isnumeric() :
            print(i)
            break
        
```

### **re.sub**

    - subtutute the string when pattern matches

re.sub(pattern, repl, string, count= 0)

    - if pattern matches with the string 
    - then replace

```python
## using the re.findall
pattern = r"\d+"

str1 = "order number 1234 shipped on 28th December 2024"

result = re.sub(pattern, "subtitute", str1)
print("original string :", str1)
print("result : ", result)
```

```python
## using the re.findall
pattern = r"\d+"

str1 = "order number 1234 shipped on 28th December 2024"

result = re.sub(pattern, "subtitute", str1, count=2)
print("original string :", str1)
print("result : ", result)
```

### **re.split**

        - the pattern \s+ 
        - \s means space
        - the split operation based on strings

```python
pattern = r"\s+"
str1 = "order number 1234 shipped on 28th December 2024"

result = re.split(pattern, str1)
print("result : ", result)
```

```python
pattern = r"\s+"
str1 = "order number 1234 shipped on 28th December 2024"

result = re.split(pattern, str1, maxsplit= 2)
print("result : ", result)
```

### ***email validation using regex***

        + email will contain 

                - digit
                - number 
                - special character (!, +, -, _, ., etc...)

*email* first part : 

            - ^ : used for start of the string 

            - [a-zA-Z0-9_.+-] : pattern if any laetter in email id have this not only one character, many character for that we need to add a "+" sign at last 

            - [a-zA-Z0-9_.+-]+ 

            - till now the pattern is ^[a-zA-Z0-9_.+-]+

*email* second part : '@' symbol 

            - pattern till now ^[a-zA-Z0-9_.+-]+@
    
*email* third part : alphanum specal

            - [a-zA-Z0-9_.+-] : pattern if any laetter in email id have this not only one character, many character for that we need to add a "+" sign at last 
            
            - [a-zA-Z0-9_.+-]+ 
            
            - till now the pattern is ^[a-zA-Z0-9_.+-]+@[a-zA-Z0-9_.+-]+ 

*email* fourth part : dot (.)

            - \ will treat dot(.) a normal character

            - till now the pattern is ^[a-zA-Z0-9_.+-]+@[a-zA-Z0-9_.+-]+\.
        
*email* fifth part : alphabet

            - [a-zA-Z]{2, }

            - {} used as group 

            - {2, } will limit values to at least 2 characters to as many

            - till now the pattern is ^[a-zA-Z0-9_.+-]+@[a-zA-Z0-9_.+-]+\.[a-zA-Z]{2, }

*email* sixth part : $ (ending)

            - $ ending of the pattern 


                                                ++++ final pattern is  "^[a-zA-Z0-9_.+-]+@[a-zA-Z0-9_.+-]+\.[a-zA-Z]{2, }$"

            

```python
pattern = r"[a-zA-Z0-9_.+-]+@[a-zA-Z0-9]+\.[a-zA-Z]{2,}"
emails = "pratham@gmail.com, kunal24@gmail.com, ranjan_23@gmial.oh, narayanRaj.23_personal@outlook.com"

result = re.findall(pattern, emails)
print("result : ", result)
# result.group()
```
