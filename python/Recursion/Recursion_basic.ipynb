#### WAP to find the multiplication of two positive number using resursion

```python
def multi(a, b) :
    if b == 0 :
        return 0
    return a + multi(a, b-1)

print("Enter two values for multiplication ")
a = int(input("enter value of a : "))
b = int(input("enter value of b : "))

print("multiplication of a and b is : ", a, "x", b, "=", multi(a, b))

```

#### WAP to perform a division of two positive number 


```python
def div(a, b) :
    if a == 0 :
        return 0
    elif b == 0 :
        return "Infinite"
    else :
        return  1 + div(a-b, b)
        

print("Enter two values for Division  ")
a = int(input("enter value of a : "))
b = int(input("enter value of b : "))

print("Division of a and b is : ", a, "/", b, "=", div(a, b))
```

#### WAP to find out the a power b.

```python
def pow(a, b) :
    if b == 0 :
        return 1
    elif a == 0 :
        return 0
    return a * pow(a, b-1)

print("Enter two values for power  ")

a = int(input("enter value of a : "))
b = int(input("enter value of b : "))

print("a power b is : ", a, "^", b, "=", pow(a, b))
```

#### Another Approach 

```python
def pow(a, n) :
    if n == 0 :
        return 1
    elif a == 0 :
        return 0

    if n % 2 == 0 : 
        return pow(a, n//2) * pow(a, n//2) 
    elif n % 2 != 0 :
        return pow(a, n//2) * pow(a, n//2) * a

print("Enter two values for power  ")

a = int(input("enter value of a : "))
b = int(input("enter value of b : "))

print("a power b is : ", a, "^", b, "=", pow(a, b))
```

#### Sum of first n number using recursion 


```python
def sumN(n) :
    if n == 0 :
        return 0
    return n + sumN(n-1)

num = int(input("enter no : "))

print("Sum till number from 0 : ", sumN(num))
```

#### Sum of Digit using recursion 

```python
def sumOfDigit(n) :
    
    if n == 0 :
        return 0
    return (n%10) + sumOfDigit(n//10)

num = int(input("enter any number  : "))
print("sum of digit = ", sumOfDigit(num))
```

#### Fibonacci series 0, 1, 1, 2, 3, 5, 8, 13...... using recursion

```python
def fibo(n) :
    
    if n == 1 :
        return 0 
    elif n == 2 :
        return 1
        
    return fibo(n-1) + fibo(n-2)

num = int(input("enter nth postition for the value : "))
print("nth position will be = ", fibo(num))
```

```python
def trace(n) :
    if n <= 0 :
        return 1 
    else : 
        return trace(n-1) + trace(n-2)
print(trace(6))
```

## Return reverse of a number of string : 

```python
def rev(s) :
    if len(s) <= 1 : 
        return s
    return rev(s[1:]) + s[0]


s = input("enter anything you want to reverse : ")
rs = rev(s)
print(rs)
```
