# Q3

```python
def sample(x):
    if x==0 or x==2:
        return 1
    return sample(x-1)*(x-1)
sample(4)

# OUTPUT --> 6
```

# Q4

```python
def f(n,r=[0]):
    if n<=0:
        return 1
    r[0]=n
    return f(n-1,r)+r[0]
f(5)

# OUTPUT --> 6
```

# Q5

```python
def howmanydash(n):
    if n==0:
        print("+")
    elif n==1:
        print("-")
    else:
        howmanydash(n-1)
        howmanydash(n-2)

        
howmanydash(5)

# OUTPUT --> -  +  -  +  -
```

# Q7. Consider the following function definition.

```python
def greet(n):
    if n > 0:
        print('hello', end=' ; ')
        greet(n - 1)
    print("world", end='--> ')

greet(6)
# OUTPUT --> hello hello hello hello hello hello world world world world world world world
```

# Q8 Consider the program given in below question, f1(8) return the values

```python
def f1(n):
    if n == 0 or n == 1:
        return n
    else:
        return 2 * f1(n - 1) + 3 * f1(n - 2)

f1(8)

# OUTPUT --> 128
```

# Q9

```python
def f(n):
    if n <= 1:
        return 1
    if n % 2 == 0:
        return f(n // 2)
    return f(n // 2) + f(n // 2 + 1)

print(f(11))

# OUTPUT --> 5
```

```python
def f(a, n):
    if n <= 0:
        return 0
    elif a[0] % 2 == 0: # Check if the first element is even
        return a[0] + f(a[1:], n - 1)
    else:
        return a[0] - f(a[1:], n - 1)

a = [12, 7, 13, 4, 11, 6]
print(f(a, len(a)))

# Output --> wrong answer -6
```
