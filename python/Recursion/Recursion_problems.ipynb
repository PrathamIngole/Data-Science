# 21. Binary search: Write a recursive function to perform binary search on a sorted array.

```python
def binary_search_recursion(l, ele, start=0, end=None) :

    if end == None :
        end = len(l) -1

    if start >= end :
        if l[start] == ele :
            return start
        else :
            return "--- Element not found ---"

    mid = (start+end)//2

    if l[mid] == ele :
        return mid
    elif l[mid] > ele :
        return binary_search_recursion(l, ele, start, mid)
    else :
        return binary_search_recursion(l, ele, mid+1, end)

#    0, 1,  2,  3,  4,   5,   6
l = [3, 6, 12, 63, 555, 853, 999]
print("postion of that element is : ", binary_search_recursion(l, 63))
```

```python
def union_of_array(l1, l2, res=None, i = 0, j =0) :

    if res == None :
        res = []

    if i == len(l1) and j == len(l2) :
        return res
    
    if i < len(l1) :
        if l1[i] not in res :
            res += [l1[i]]
        return union_of_array(l1,l2,res, i+1, j)

    if j < len(l2) :
        if l2[j] not in res : 
            res += [l2[j]]
        return union_of_array(l1, l2, res, i, j+1) 
    

n = int(input("enter size of 1st array : "))
l1 = [int(input("enter l1 element")) for i in range(n)]
print("Entered first array is : ", l1)

m = int(input("enter size of 2nd array : "))
l2 = [int(input("enter l1 element")) for i in range(m)]
print("Entered second array is : ", l2)

print("union of the two arrays is : ", union_of_array(l1, l2))
```

# 50. Find all permutations of a set of numbers without repetition: Write a recursive function to generate all permutations of a set of numbers without repetition.

```python
def permutation(n, ans= None) :
    if ans == None :
        ans = ""
    if len(n) == 0 :
        print(ans) 
        return
    for i in range(len(n)) :
        permutation(n[:i] + n[i+1:], ans+n[i])

n = input("enter any number :")
permutation(n)
```
