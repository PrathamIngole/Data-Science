# 1.find the output

```python
def f(n):
    if n <= 1:
        return 1
    if n % 2 == 0:
        return f(n // 2)
    return f(n // 2) + f(n // 2 + 1)

print(f(11))
# OUTPUT --> 5
```

# 2.find the output

```python
def f(n):
    i = [1]
    if n >= 5:
        return n
    n = n + i[0]
    i[0] += 1
    return f(n)

print(f(1))
# OUTPUT --> 5
```

# 3.find the output

```python
def function(num):
    if num > 0:
        function(num - 1)
        print(num - 1, end='  ')
        function(num - 1)

function(3)

# OUTPUT --> 0  1  0  2  0  1  0
```

# 4.find the output

```python
def fun1(n):
    if n == 0:
        return
    print(n, end='  ') # Printing without a newline
    fun2(n - 2)
    print(n, end='  ')

def fun2(n):
    if n == 0:
        return
    print(n, end='  ')
    fun1(n + 1)
    print(n, end='  ')

fun1(3)

# OUTPUT --> 3  1  2  2  1  3
```

# 5. find the output

```python
def func(n):
    if n < 3:
        return 1
    else:
        return func(n - 1) + func(n - 3) + 1

print(func(5))

# OUTPUT --> 7
```

# 6.find the output

```python
def foo(n, r):
    if n > 0:
        return (n % r) + foo(n // r, r)
    else:
        return 0

print(foo(513, 2))

# reccursion call --> 1 + foo(256, 2) --> 0 + foo(128, 2) --> 0 + foo(64, 2) --> 0 + foo(32, 2) --> 0 + foo(16, 2) --> 0 + foo(8, 2) --> 0 + foo(4, 2) --> 0 + foo(2, 2) --> 0 + Foo(1, 2) --> 1 + foo(0, 2) --> 0
# OUTPUT --> 2
```
