```python
# creation of Node 

class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None
```

```python
# Printing of the linked list 

def printLL(head) :
    if head == None :
        print("list is empty") 
        return
    
    temp = head 
    
    while temp != None :
        print(temp.data, end=" ==> ") 
        temp = temp.next
    print(None)
```

```python
# searching in Linked List

def search_1(head, data) :
    if head == None :
        print("list is empty") 
        return head
    
    if head.data == data :
        return 1
    
    pos = 1
    temp = head

    while temp != None:
        if temp.data == data :
            return pos
        pos += 1
        temp = temp.next
        
    return "element not found"
```

```python
head = Node(10)
head.next = Node(11)
head.next.next = Node(12)
head.next.next.next = Node(13)
head.next.next.next.next = Node(14)
printLL(head)
```

```python
printLL(head)
pos = search_1(head, 10)
print("serched element in position : ", pos)
```

```python
def search_2_through_reccursion(head, data, pos = 1) :

    if head == None :
        return "element not found"
    
    if head.data == data :
        return f"elemtent found at postion : {pos}"

    return search_2_through_reccursion(head.next, data, pos+1)
```
