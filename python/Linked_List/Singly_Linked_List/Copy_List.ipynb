```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None
```

```python
def printLL(head) :
    if head == None :
        print("Linked List is empty")
        return head
    
    temp = head 
    while temp != None :
        print(temp.data, end=" ==> ")
        temp = temp.next

    print(None)
    
```

```python
def Insert_at_last(head, data) :
    if head == None :
        return Node(data)
    
    if head.next == None :
        head.next = Node(data) 
        return head
    
    temp = head

    while temp.next != None :
        temp = temp.next
    temp.next = Node(data)

    return head
```

```python
def Copylist(head) :
    if head == None:
        return head
    head2 = Node(head.data)
    if head.next == None :
        return head2
    
    temp = head 
    new = head2


    while temp.next != None :
        new.next = Node(temp.next.data)
        temp = temp.next
        new = new.next

    return head2
```

```python
head = None
head = Insert_at_last(head, 10)
head = Insert_at_last(head, 16)
head = Insert_at_last(head, 6)
head = Insert_at_last(head, 22)
head = Insert_at_last(head, 34)
head = Insert_at_last(head, 76)
printLL(head)
new = Copylist(head)
print("our original list before copying new list ")
printLL(head)
print("our copied list after copying new list ")
printLL(new)
print(new == head)

print("\n\n")
new = Insert_at_last(new, 100)
head = Insert_at_last(head, 200)
head = Insert_at_last(head, 300)
head = Insert_at_last(head, 500)
print("our head list : ")
printLL(head)
print("our new list : ")
printLL(new)
print(new == head)
```
