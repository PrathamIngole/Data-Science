```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None
```

```python
def printLL(head) :
    if head == None :
        print("Linked List is empty")
        return head
    
    temp = head 
    while temp != None :
        print(temp.data, end=" ==> ")
        temp = temp.next

    print(None)
    
```

```python
def Insert_at_last(head, data) :
    if head == None :
        return Node(data)
    
    if head.next == None :
        head.next = Node(data) 
        return head
    
    temp = head

    while temp.next != None :
        temp = temp.next
    temp.next = Node(data)

    return head
```

```python
def copyList(head) :
    if head == None :
        return head 
    
    head2 = Node(head.data)
    new = head2
    temp = head 
    while temp.next != None :
        new.next = Node(temp.next.data)
        temp = temp.next
        new = new.next

    return head2
```

```python
def reverseList(head) :
    if head == None :
        return None 
    if head.next == None :
        return head
    
    curr = head
    prev = None
    while curr != None :
        temp = curr.next
        curr.next = prev
        prev = curr
        curr = temp
    return prev
```

```python
def checkPalindrome(head) :
    if head == None :
        return False
    if head.next == None :
        return True
    
    newList = copyList(head) 
    newList = reverseList(newList)

    temp1 = head
    temp2 = newList

    while temp1 != None and temp2 != None :
        if temp1.data != temp2.data :
            return False
        temp1 = temp1.next
        temp2 = temp2.next

    return True
```

```python
head = None
head = Insert_at_last(head, 10)
head = Insert_at_last(head, 20)
head = Insert_at_last(head, 20)
head = Insert_at_last(head, 10)
printLL(head)


new = copyList(head)

print("\nold List : ")
printLL(head)

print("\nnew list :  ")
printLL(new)

print("\n")
print(head == new)

print("\n\nchecking is list is palindrome or not")
printLL(head)
print("\nis palindrome : ", checkPalindrome(head))

```
