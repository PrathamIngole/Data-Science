```python
# creation of Node 

class Node :
    
    def __init__(self, data) :
        self.data = data
        self.next = None
```

```python
# Printing of the linked list 

def printLL(head) :
    if head == None :
        print("list is empty") 
        return
    
    temp = head 
    
    while temp != None :
        print(temp.data, end=" ==> ") 
        temp = temp.next
    print(None)
```

```python
def countNode(head) :

    if head == None :
        return 0
    
    count = 0
    temp = head

    while temp is not None :
        count += 1
        temp = temp.next

    return count
```

```python
def Insert_ele_sorted(head, data) :

    if head == None :
        return Node(data) 
    
    node = Node(data)
    if head.data > data :
        node.next = head
        return node
    
    temp = head
    while temp.next != None and temp.data <= data:
        if temp.next.data > data:
            node.next = temp.next
            temp.next = node
            return head
        else :
            temp = temp.next
    temp.next = node
    return head
```

```python
def nth_node_form_begining(head, n) :
    if head == None :
        print("list is empty")
        return head
    
    if n <= 1 :
        return head.data
    
    pos = 1
    temp = head
    while pos < n and temp.next != None:
        temp = temp.next
        pos += 1

    if temp == None :
        print("position invalid")
        return None
    
    return temp.data
```

```python
def nth_node_form_end(head, n) :
    if head == None :
        print("list is already empty")
        return head

    count = countNode(head)
    if n == count :
        return head.data
    
    if n > count :
        print("position invalid")
        return None

    pos = 1 
    temp = head 
    while pos < count - n + 1 and temp.next != None :
        temp = temp.next
        pos += 1
    
    return temp.data
```

```python
head = Node(10) 
head = Insert_ele_sorted(head, 20)
head = Insert_ele_sorted(head, 30)
head = Insert_ele_sorted(head, 40)
head = Insert_ele_sorted(head, 50)
```

```python
printLL(head)
```

```python
node = nth_node_form_begining(head, 8)
print(node)
```

```python
node = nth_node_form_end(head, 5)
print(node)
```
