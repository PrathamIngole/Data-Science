```python
# creation of Node 

class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None
```

```python
# Printing of the linked list 

def printLL(head) :
    if head == None :
        print("list is empty") 
        return
    
    temp = head 
    
    while temp != None :
        print(temp.data, end=" ==> ") 
        temp = temp.next
    print(None)
```

```python
def Insert_ele_sorted(head, data) :

    if head == None :
        return Node(data) 
    
    node = Node(data)
    if head.data > data :
        node.next = head
        return node
    
    temp = head
    while temp.next != None and temp.data <= data:
        if temp.next.data > data:
            node.next = temp.next
            temp.next = node
            return head
        else :
            temp = temp.next
    temp.next = node
    return head
```

```python
def removeDuplicates(head) :
    if head == None :
        print("no duplicates present")
        return head
    
    temp = head 
    while temp != None and temp.next != None :
        if temp.data == temp.next.data :
            temp.next = temp.next.next
        else :
            temp = temp.next
    return head
```

```python
head  = None
head = Insert_ele_sorted(head, 8)
head = Insert_ele_sorted(head, 15)
head = Insert_ele_sorted(head, 27)
head = Insert_ele_sorted(head, 40)
head = Insert_ele_sorted(head, 69)
head = Insert_ele_sorted(head, 10)
head = Insert_ele_sorted(head, 50)
head = Insert_ele_sorted(head, 20)
head = Insert_ele_sorted(head, 20)
head = Insert_ele_sorted(head, 20)
head = Insert_ele_sorted(head, 27)
printLL(head)
```
