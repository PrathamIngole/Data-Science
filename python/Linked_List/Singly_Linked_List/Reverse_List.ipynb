```python
# creation of Node 

class Node :
    
    def __init__(self, data) :
        self.data = data
        self.next = None
```

```python
# Printing of the linked list 

def printLL(head) :
    if head == None :
        print("list is empty") 
        return
    
    temp = head 
    
    while temp != None :
        print(temp.data, end=" ==> ") 
        temp = temp.next
    print(None)
```

```python
def Insert_ele_sorted(head, data) :

    if head == None :
        return Node(data) 
    
    node = Node(data)
    if head.data > data :
        node.next = head
        return node
    
    temp = head
    while temp.next != None and temp.data <= data:
        if temp.next.data > data:
            node.next = temp.next
            temp.next = node
            return head
        else :
            temp = temp.next
    temp.next = node
    return head
```

```python
def reverseList(head) :
    curr = head 
    prev = None
    while curr != None :
        temp = curr.next
        curr.next = prev
        prev = curr
        curr = temp
    return prev
```

```python
head = Node(10) 
head = Insert_ele_sorted(head, 854)
head = Insert_ele_sorted(head, 74)
head = Insert_ele_sorted(head, 56)
head = Insert_ele_sorted(head, 3)
head = Insert_ele_sorted(head, 65)
printLL(head)
head = reverseList(head)
printLL(head)
```
