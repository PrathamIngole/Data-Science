```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None
```

```python
def printLL(head) :
    if head == None :
        print("Linked List is empty")
        return head
    
    temp = head 
    while temp != None :
        print(temp.data, end=" ==> ")
        temp = temp.next

    print(None)
    
```

```python
def Insert_at_last(head, data) :
    if head == None :
        return Node(data)
    
    if head.next == None :
        head.next = Node(data) 
        return head
    
    temp = head

    while temp.next != None :
        temp = temp.next
    temp.next = Node(data)

    return head
```

```python
def check_equals(head1, head2) :
    if head1 == None and head2 == None :
        print("both linked list are empty")
        return True
    temp1 = head1
    temp2 = head2
    while temp1 != None and temp2 != None :
        if temp1.data != temp2.data :
            return False
        temp1 = temp1.next
        temp2 = temp2.next

    if temp1 == None and temp2 == None :
        return True
    else :
        return False
```

```python
head1 = None
head1 = Insert_at_last(head1, 10)
head1 = Insert_at_last(head1, 20)
head1 = Insert_at_last(head1, 30)
head1 = Insert_at_last(head1, 40)
head1 = Insert_at_last(head1, 50)
head1 = Insert_at_last(head1, 60)
print("list 1 : ")
printLL(head1)

head2 = None
head2 = Insert_at_last(head2, 10)
head2 = Insert_at_last(head2, 20)
head2 = Insert_at_last(head2, 30)
head2 = Insert_at_last(head2, 40)
head2 = Insert_at_last(head2, 50)
head2 = Insert_at_last(head2, 60)
print("\nlist 2 : ")
printLL(head2)

check =  check_equals(head1, head2)
print("\nlist 1 and list 2 are same : ", check)

```
