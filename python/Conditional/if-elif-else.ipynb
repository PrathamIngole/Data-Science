--> checking a no is inside circle's circumference or not 

```python
r = float(input("enter the radius of circle : "))

px = int(input("enter the point x coordinate : "))
py = int(input("enter the point y coordinate : "))

d = (((px**2)+ (py**2))**0.5)
print(d)

if d > r :
    print("point is outside")
elif d < r :
    print("point is inside")
else :
    print("point is on circumference")
    
```

--> check if a number is even or not and if even then is it a multiple of 3 or not

```python
A = int(input("enter a no :  "))

if A & 1 == 0 :
    if A % 3 == 0 : 
        print("its even and multiple of 3")
    else :
        print("its even but not multiple of 3 ")
else :
    print("number is not even")
```
