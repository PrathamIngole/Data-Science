## Insertion sort :

                - insertion sort is sorting method which is used to sort an unsorted array 
                - in insertion sort we select the element and insert it on the position it belong
            

```python
# defining one funciton for sortion applying insertion sort logic

def InsertionSort(arr) :
    print("size of your array : ", len(arr))
    print("your array is : ", arr)

    # iterating array from 1st index instead of 0th index so that we select arr[1] element and find its position behind it 
    # and insert it if it belongs to there
    for i in range(1, len(arr)) :
        # selecting one element
        ele = arr[i]

        # taking previous array by selecting array from i-1 to 0 index
        j = i-1

        # place where you find element's position
        while j >= 0 and ele < arr[j] :
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = ele

    # return if you want to give a sorted iterator object
    return arr

    # else comment **return arr**
```

```python
# dynamically taking size of array also inserting input elements of user choice
arr = [int(input(f"enter {i+1}th element : ")) for i in range(int(input("enter size of array : ")))]

# taking sorted array as result 
result = InsertionSort(arr)

# printing result
print("result is : ", result)

# checking if array element is sorted in place 
print("original array after operation is : ", arr)
```
