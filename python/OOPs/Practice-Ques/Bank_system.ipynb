```python
# imported datetime to get the current date and time for track of transactions
from datetime import datetime

class Bank :

    # created a account dictionary which stores different users data all together same for all instances 
    # private class attribute 
    __accounts = {}

    # structure of __accounts :
    # __accounts = {
    #     Acc_no : {
    #         "password" : str,
    #         "balance" : float,
    #         "transaction" : []
    #     },
    #     another account....
    # }


    # for current instance we need to set attributes for all methods
    def __init__(self, acc_no, password):
        self.Acc_no = acc_no
        self.Password = password


        # authenticate first if the user exist or not
        if self.authenticate() :
            print("Login Successful")
            print("Available Options : ")
            print("1. check balance    2. credit amount    3. debit amount")
        else : 
            print("account not exist.... or may be wrong password ")
            print("please create account using : create_account()")


    # authenticate function check if the person entered correct inputs or not and if one of the field is correct
    def authenticate(self) : 
        return (
            self.Acc_no in Bank.__accounts and
            Bank.__accounts[self.Acc_no]["password"] == self.Password
        )
    

    # creating account 
    def create_account(self) :
        if self.Acc_no in Bank.__accounts :
            print("Account already exist....")
            return
        else : 
            Bank.__accounts[self.Acc_no] = {
                "password" : self.Password,
                "balance" : 0,
                "transactions" : []
            }
            print("Account Created successfully....")

    
    # ----------------credit amount-----------------
    def credit(self, amount) :
        if self.authenticate() :
            self.__transaction("Credited", amount)
            Bank.__accounts[self.Acc_no]["balance"] += amount
            print(f'\nAmount of rupees {amount} is Credited on {Bank.get_date_time()}')
            print(f"From Account no. {self.Acc_no}")
            print(f"Updated Balance is rupees {Bank.__accounts[self.Acc_no]["balance"]}/-\n")
        else : 
            print("Authentication Failed")
            return
        

    # ----------------debit amount-----------------
    def debit(self, amount) :
        if self.authenticate() :
            if amount > Bank.__accounts[self.Acc_no]["balance"] :
                print("Insufficient funds.....")
                return 
            else :
                self.__transaction("Debited", amount)
                Bank.__accounts[self.Acc_no]["balance"] -= amount
                print(f'\nAmount of rupees {amount} is Debited on {Bank.get_date_time()}')
                print(f"From Account no. {self.Acc_no}")
                print(f"Updated Balance is rupees {Bank.__accounts[self.Acc_no]["balance"]}/-\n")
        else :
            print("Authentication Failed")
            return 
        
    
    # ------------------------ check balance ----------------
    def check_balance(self) :
        if self.authenticate() :
            print(f"On {self.get_date_time()}")
            print(f"Balance in Account no. {self.Acc_no} is rupees {Bank.__accounts[self.Acc_no]["balance"]}/-")
            print()
        else : 
            print("Authentication Failed")
            return
        

    # ------------- get transaction history for the current user ----------------------------
    def transaction_history(self) :
        if self.authenticate() :
            print(f"transaction history for Account no. {self.Acc_no}\n")
            if len( Bank.__accounts[self.Acc_no]["transactions"]) == 0 :
                print("no transactions yet")
            for i in Bank.__accounts[self.Acc_no]["transactions"] :
                print(i)
            print(f"\nCurrent Balance : {Bank.__accounts[self.Acc_no]["balance"]}\n")
        else :
            print("Authentication Failed")
            return 


    # ----------------- add transactions -----------------------
    # private class method
    def __transaction(self, type, amount) :
        self.entry = f"{type} {amount} on {self.get_date_time()}"
        Bank.__accounts[self.Acc_no]["transactions"].append(self.entry)

    
    # defined a static method because it does not related to any instance, it is independent 
    @staticmethod
    def get_date_time() :
        return datetime.now().strftime("%d-%m-%Y  %H:%M:%S")
```

```python
person1 = Bank(1223545, "@1234")
```
