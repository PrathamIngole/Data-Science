## Create a circle class to create a circle with radius r using the constructor. Define an Area() method of the class which calculates the area of the circle. Define a Perimeter() method of the class which allows you to calculate the Parimeter of the circle.

```python
class Circle :
    def __init__(self, radius):
        self.radius = radius
        print("Circle Created")
    
    
    def Area(self) :
        return 3.14*(self.radius**2)

    def Perimeter(self) :
        return  round(2*3.14*self.radius, 2)
```

```python
c1 = Circle(5)
```

```python
area = c1.Area()
perimeter = c1.Perimeter()
```

```python
print("area of circle is : ", area)
print("Perimeter of a circle : ", perimeter)
```
