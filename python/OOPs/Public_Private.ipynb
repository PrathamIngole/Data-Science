# Public & Private :

    - public attributes and methods can be access outside the class scope.
    - private attributes and methods cann't be access outside class scope. it can only be available inside the class scope.


```python
import datetime

class Account:

    def __init__(self, acc_no, balance, password) : 
        # public attribute
        self.Acc_no = acc_no
        self.Balance = balance
        # private attribute
        self.__Pass = password

    def __get_acc_details(self) :
        print(f"account {self.Acc_no} has {self.__Pass} password and {self.Balance} balance")

    def details(self) :
        self.__get_acc_details()

    @staticmethod
    def __get_date() :
        return datetime.date.today()
```

```python
person1 = Account(1234, 500, "@123")
```

```python
person1.__get_date()
# error cause accessing private method
```

```python
person1.Balance
# public attribute
```

```python
person1.__Pass
# private attribute
```

```python
person1.details()
# public method 
# here one intersting thing that we are accessing private attributes and method under the public method which again in class scope.
```
