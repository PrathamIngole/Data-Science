# Pillars of OOP's :
            1. Abstraction : 
                - hiding the implementation details of a class and only showing essential features to the user.

            2. Encapsulation : 
                - Wrapping data and functions into a single unit(object) called Encapsulation.

            3. Inheritance : 
                - When one class(child/derived) derieves the properties & methods of another class(parent/base).

                                + types of Inheritance : 
                                            1. Single Inheritance.
                                            2. Multi-level Inheritance.
                                            3. Multiple Inheritance.
                                        
            4. Polymorphism : 
                - defining different meaning of the operators for different operations.
                - dender function are the real use here.

```python
class Car:
    def __init__(self) :
        self.acc = False
        self.brk = False
        self.clutch = False
    
    def start(self) :
        self.clutch = True
        self.acc = True
        print("car started ... ")

    def stop(self) :
        self.brk = True
        self.acc = False
        print("car stopped ...")
```

## Abstraction :

```python
car1 = Car()
car1.start()
car1.stop()
```

    - here we hide all the unnesesary data like clutch break functioning cause driver doesn't need to see that.. 
    - driver only want to know is car started or not, is car stopped or not.

## Encapsulation :

    - whole class code is Encapsulation cause we organised class and all related data/attributes-method in one single place.
    - i.e. Car class

## Inheritance :

    - aquiring properties & methods of the parent class and use it on its own scope is Inheritance.

            + types of Inheritance.

                    1. Single Inheritance : 
                                    - child class inherits the properties of the parents class.
                                

```python
class car :

    @staticmethod
    def start() :
        print("car started")

    @staticmethod
    def stop() :
        print("car stoped")

class toyotaCar(car) :

    def __init__(self, name):
        self.name = name    
```

```python
car1 = toyotaCar("Fortuner")
print(car1.name)
car1.start()
car1.stop()
```

                    2. Multi-level Inheritance :
                                    - child class inherits the properties of the base class and then so on.

```python
class A:
    varA = "welcome A"

class B(A) :
    varB = "welcome B"
    def __init__(self, brand) :
        self.name = brand
        print(self.name)
class C(B) :
    varC = "welcome to C"


c1 = C("naam rakh le")

print(c1.varA)
print(c1.varB)
print(c1.varC)
```

                    3. Multiple Inheritance : 
                                - multiple parents class are inherited into a single base class.

```python
class A :
    varA = "hey class A"
    def __init__(self):
        print("class A")

class B :
    varB = "hey class B"
    def __init__(self):
        print("calss b", varB)

class C(A, B) :
    varC = "hey class C"

c1 = C()
c1.varA
```

```python
c1.varB
```

```python
c1.varC
```

                        4. Polymorphism : 
                            - one operation/operator different functionality.
                        

```python
class Complex :
    def __init__(self, real, img):
        self.real = real 
        self.img = img
        print(real,"i +", img, "j")
    

    #this is a dunder function __add__there are more like this.
    def __add__(self, num2) :
        newReal = self.real + num2.real 
        newImg = self.img + num2.img
        return Complex(newReal, newImg)
```

```python
num1 = Complex(3, 5)
num2 = Complex(5, 7)
```

```python
num3 = (num1 + num2)
```
