### Default Constructor :

                - when class is created and we define all the attributes and methods inside the class.
                - then we create some objects/Instance of the class.
                - In the time of creating a object a default method is created which is called the __init__ method :

        

```python
class Student :
    def __init__ (self) :
        print(self, "\nconstructed created when the object is created") 

    College_Name = "ABC college"

s1 = Student()
```

- there you  can see one reference is called automatically without calling anything.
- we just created a Student object in that time pyhton automatic define a constructor and called it.
- this is called Default Constructor.



## Parameterized Constructor :

        - when we define constructor manually with our need operations and initializations then it will be called as a parameterized constructor.

        - whenever we pass parameters/data in the time of object creation to use that data and initialise that data then it is called parameterized constructor.

```python
class Student_marks:
    def __init__ (self, name, marks) :
        self.name = name
        self.marks = marks

    subjects = ['maths', 'physics', 'chemistry']

    def print(self) :
        print(f"hey!! {self.name} your marks is : ")
        for sub, mark in zip(self.subjects, self.marks) :
            print(f"{sub} marks : {mark}")

s1 = Student_marks("Pratham", [98, 90, 93])    

print(s1.name)
print(s1.subjects)
print(s1.marks) 
s1.print()
```

+ there we defined a constructor with the arguments.
+ In time of creation of that object we pass some parameters then those parameter is used to initialize some class/Instance attributes.
