## Methods : 
        1. Instance Method - which uses self as a argument.
        2. Static Method - which is independent of the self arrgument.
        3. Class Method - which uses cls as a argument to access the class attributes.

```python
class Student :
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    subjects = ["Maths", "Chemistry", "Physics"]

    # Instance method
    def get_avg(self) :
        return sum(self.marks) / len(self.marks)
    
    def print_sub_marks(self) :
        for sub, mark in zip(Student.subjects, self.marks) :
            print(f"{sub} marks : {mark}")
    
    # Static method
    @staticmethod
    def get_sub() :
        for i in Student.subjects :
            print(i, end=", ")
        print()

    @staticmethod
    def get_grade(avg) :
        if avg > 75 :
            print(f"Student gets grade : A")
        elif avg > 65 :
            print(f"Student gets grade : B")
        else :
            print(f"Student gets grade : Ache se padh beta")

    # Class method
    @classmethod
    def change_subjects(cls, subjects) :
        cls.subjects = subjects
```
