### Input output functions : 

- calculating simple intrest by taking user input and give ans by print() fucntion.

```python
# print() functions helps us to display content as output --- it takes N numbers of parameter seperated by commas(,).

print("calculating SIMPLE INTREST")


# where input() function help us to take input from the user -- it accept string by default to convert those into desired data type use type casting 

p = int(input("enter the principle"))
r = int(input("enter the rate"))
t = int(input("enter the time"))

SI = (p*r*t)/100

print("Simple intrest : ", SI)
```

### calculating perimeter of a circle


```python
print("perimeter of a Circle")

r = float(input("enter radius"))

perimeter = 2*3.14*r

print("perimeter of a circle is : ", perimeter)
```

- calculating Area of a triangle

```python
print("Calculating Area of a triangle ===> ")

print("enter sides of triangle")
side1 = float(input("enter the side1"))
side2 = float(input("enter the side2"))
side3 = float(input("enter the side3"))

s =  (side1+side2+side3)/2

Area = (s*(s-side1)*(s-side2)*(s-side3))**0.5


print("Area of a triangle : ", Area)
```

- Distance between two coordinate : 


```python
print("Distance beteen two coordinate  ")

x1, y1 = input("enter value of x1 and y1").split()

x2, y2 = input("enter value of x2 and y2").split()

x1 = int(x1)
x2 = int(x2)
y1 = int(y1)
y2 = int(y2)

d = (((y2-y1)**2) + ((x2 - x1)**2))**0.5


print("distance is : ", d)
```
